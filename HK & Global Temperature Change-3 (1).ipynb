# Trends of Global and Hong Kong temperature change in 1990-2019

## 1.1 The overall Trend of Hong Kong temperature change in recent thirty years

```python
# import packages
import requests
import pandas as pd
import numpy as np

# import csv format data
dfHKTemp = pd.read_csv(r'/Users/hazel/Downloads/https___www.hko.gov.hk_en_cis_ (1).csv')
```

```python
# process the data of Hong Kong temperature change 
dfHKTemp
```

```python
# select relevant columns
dfHKTemp[['Year', 'Mean (deg. C)\n']]
```

```python
# assign selected columns back to the variable
dfHKTemp=dfHKTemp[['Year', 'Mean (deg. C)\n']]
```

```python
# select the data for the recent 30 years
# cleaned data
dfHKTemp30=dfHKTemp[-30:]
```

```python
# plot a regression graph to show HK's overall temeperature change 
#in the past 30 years
import matplotlib.pyplot as plt

# Convert the dataset to numpy
x = dfHKTemp30.iloc[:, 0].to_numpy()
y =  dfHKTemp30.iloc[:, 1].to_numpy()

# Create scatter plot
plt.plot(x, y, 'o')


# Plot a linear regression line, m = slope, b=intercept
m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x + b)
```

## 1.2 The trends of Hong Kong temperature change in every ten years

```python
# Assign variable of first 10 years
dfHKTempFirst10 = dfHKTemp30[0:10]
```

```python
dfHKTempFirst10
```

```python
# plot a line graph to show the HK temperature change from 1990 to 1999
temp = []
tempDf = dfHKTempFirst10['Mean (deg. C)\n'].tolist()
diff = (tempDf[9] - tempDf[0])/ 9

for i in range (10):
  if i==0:
    temp.append(tempDf[0])
  elif i==9:
    temp.append(tempDf[9] )
  else:
    temp.append(tempDf[0] + diff*i)

df = pd.DataFrame({
   'HK Temp': dfHKTempFirst10['Mean (deg. C)\n'].tolist(),
   'Average': temp
   }, index=dfHKTempFirst10['Year'].tolist())
lines = df.plot.line()

print ("Percentage Change = %.2f" %((tempDf[9] - tempDf[0])/tempDf[0]) )
```

## 3 Find out the correlation between the global trend and Hong Kong trend
